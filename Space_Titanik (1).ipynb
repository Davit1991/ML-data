```python
df_train = df_train.drop('Destination', axis=1)
df_train = df_train.drop('Name', axis=1)
```

```python
df_train = df_train.drop('Cabin', axis=1)
df_train['HomePlanet'] = df_train['HomePlanet'].replace(to_replace='Europa', value=0)
df_train['HomePlanet'] = df_train['HomePlanet'].replace(to_replace='Earth', value=1)
df_train['HomePlanet'] = df_train['HomePlanet'].replace(to_replace='Mars', value=2)
```

```python
X_train = df_train.drop('Transported', axis=1)
y_train = df_train['Transported']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
model = RandomForestClassifier()
scaler = StandardScaler()
```

```python
pipe = Pipeline([('scaler', scaler), ('rfc', model)])
```

```python
params = {
    'rfc__n_estimators':list(range(70, 75)),
    'rfc__criterion':["gini", "entropy", "log_loss"],
    'rfc__max_depth':[2, 4, 8, 32],
    'rfc__max_features':["sqrt", "log2", "auto"],
    'rfc__bootstrap':[True, False],
    'rfc__oob_score':[True, False]
}
```

```python
full_model = GridSearchCV(pipe, param_grid=params, cv=3, verbose=2, scoring='accuracy')
```

```python
full_model.fit(X_train, y_train)
```

```python
full_model.cv_results_
```

```python
full_model.best_params_
```

```python
y_pred_val = full_model.predict(X_val)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_pred_val, y_val))
```
